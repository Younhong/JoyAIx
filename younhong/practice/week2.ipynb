```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def mileToKm(x):
    return 1.61 * x
def plotMileToKm(x, y):
    plt.figure()
    plt.plot(x, y, 'ob') #-sr: 빨간 줄로 점을 잇는다.
    plt.title("Mile to km")
    plt.xlabel("Mile")
    plt.ylabel("Km", fontsize=20)
    plt.show()
```

```python
for mile in range(0, 5):
    print('{}mi:{}km'.format(mile, mileToKm(mile)))
```

```python
x = [0,1,2,3,4]
y = [0, 1.61, 3.22, 4.83, 6.44]
plotMileToKm(x, y)
```

```python
y = []
for mile in range(0, 5):
    y.append(mileToKm(mile))
print(y)
```

```python
y = [mileToKm(mile) for mile in range(0, 5)]
print(y)
```

```python
x = np.arange(0, 5)
y = mileToKm(x)
print(x)
print(y)
plotMileToKm(x, y)
```

```python
n = 1000000
w = [np.random.random() for _ in range(n)]
x = [np.random.random() for _ in range(n)]

wnum = np.array(w)
xnum = np.array(x)
```

```python
%%timeit
total = 0
for i in range(n):
    total += w[i]*x[i]
print(total)
```

```python
%%timeit
total = np.dot(wnum, xnum)
print(total)
```

```python
# 배열 속성을 출력하는 함수
def pprint(arr):
    print("type:{}, size:{}".format(type(arr), arr.size))
    print("shape:{}, ndim/rank:{}, dtype:{}".format(arr.shape, arr.ndim, arr.dtype))
    print("Array's Data:")
    print(arr)
```
